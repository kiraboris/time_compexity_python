**Complexity basics**
-----

**Problem:**  
We want to test execution time for functions _depending on a variable **n**_.

```python
def my_func_1(n):
    for _ in range(0, n): 
        pass
```

```python
def my_func_2(n):
    for _ in range(0, n):
        for _ in range(0, n):
            pass
```

What time dependence on **n** would you expect? :)  

---
Plotting helper:

```python
import matplotlib.pyplot as plt
import numpy as np
import timeit
import time


def my_timeit(test_func, n, T):
    """
        test_func: callable to test
        T: number of test runs
    """
    results = []
    for _ in range(0, T):
        
        t1 = time.process_time()
        test_func(n)   # This runs the test 
        t2 = time.process_time()
        
        results.append((t2-t1))
    return np.mean(results)*1e3


def plot_time_vs_n(test_func, n_max, T, **kwargs):
    """
        test_func(n): callable to test
        
        n_max: 
               to investigate dependence on n, we use
               range(start=n_max/10, stop=n_max, step=n_max/100)
               
        T: number of test runs per data point 
        
        **kwargs: passed to plt.plot (plot color etc.)
    """  
    xxx = []
    yyy = []
    for n in range(n_max//10, n_max, n_max//100):
        xxx.append(n)
        yyy.append(my_timeit(test_func, n, T))
    
    plt.xlabel("n")
    plt.ylabel("time / msec")
    return plt.plot(xxx, yyy, **kwargs)
```

```python
plot_time_vs_n(my_func_1, n_max=1000, T=200, color='blue');
```

```python
plot_time_vs_n(my_func_2, n_max=1000, T=10, color='red');
```

```python
plot_time_vs_n(my_func_1, n_max=1000, T=10, color='blue');
plot_time_vs_n(my_func_2, n_max=1000, T=10, color='red');
```

Conclusion
---

**Time complexity** of a function by a variable **n** is the dependence of its execution time on **n**.<br>

**"Big-O"** notation: 
<ul>
    <li>O(n) -- linear,  
    <li>O(n\*n) -- quadratic,  
    <li>etc...
</ul>
