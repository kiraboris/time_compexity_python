**Comparing two lists**
-----

**Problem:**  
Find all those transition by their quantum number _J_ from one list, which are also present in another list.  
Let the lists be _list_a_ == [...], _list_b_ == [...], len(_list_a_) = len(_list_b_) = _n_.
  
**Naive idea:**  
Loop through both lists and compare.
  
**Hashing:**  
Convert one of the lists into a set.

```python
import random

class BaseQuantumListTest:
    def __init__(self, n):
        self.lower = 0      # J = 0
        self.higher = 200   # J = 200
        self.list_a = [random.randint(self.lower, self.higher) for _ in range(0, n)]
        self.list_b = [random.randint(self.lower, self.higher) for _ in range(0, n)]
        self.set_b = set(self.list_b)
        
class NaiveCompareTest(BaseQuantumListTest): 
    def run(self):
        common_j = []
        for j in self.list_a:
            if j in self.list_b:    # implicit loop in list_b
                common_j.append(j)
                
class HashedCompareTest(BaseQuantumListTest): 
    def run(self):
        common_j = []
        for j in self.list_a:  
            if j in self.set_b:     # lookup in set_b
                common_j.append(j)
            
```

```python
import matplotlib.pyplot as plt
import numpy as np
import timeit
import time


def my_timeit_with_setup(TestClass, n, T):    
    """
        TestClass: class[name] of test case
        T: number of test runs
    """
    results = []
    for _ in range(0, T):
        test = TestClass(n)  
        t1 = time.process_time()
        test.run()        
        t2 = time.process_time()
        results.append((t2-t1))        
    
    return np.mean(results)*1e6


def plot_time_vs_n(TestClass, n_max=1000, T=10, **kwargs):
    """
        TestClass(n): class[name] of test case
        
        n_max: 
               to investigate dependence on n, we use
               range(start=n_max/10, stop=n_max, step=n_max/100)
               
        T: number of test runs per data point 
        
        **kwargs: passed to plt.plot (plot color etc.)
    """  
    xxx = []
    yyy = []
    for n in range(n_max//10, n_max, n_max//100):
        xxx.append(n)
        yyy.append(my_timeit_with_setup(TestClass, n, T))
    
    plt.xlabel("n")
    plt.ylabel("time / usec")
    return plt.plot(xxx, yyy, **kwargs)
```

```python
plot_time_vs_n(NaiveCompareTest, color='blue');
```

```python
plot_time_vs_n(HashedCompareTest, color='red');
```

```python
plot_time_vs_n(NaiveCompareTest, color='blue');
plot_time_vs_n(HashedCompareTest, color='red');
```
